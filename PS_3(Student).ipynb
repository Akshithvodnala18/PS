```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/student ')

# Task 1: Histogram Analysis
plt.figure(figsize=(10, 6))
sns.histplot(df['GPA'], bins=10, kde=True)
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()

# Analyze the shape of the distribution
print("The shape of the GPA distribution can be analyzed by observing the histogram above.")

# Task 2: Scatterplot Exploration
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Attendance Rate (%)', y='GPA', data=df)
plt.title('GPA vs. Attendance Rate')
plt.xlabel('Attendance Rate (%)')
plt.ylabel('GPA')
plt.show()

# Add a trendline
sns.lmplot(x='Attendance Rate (%)', y='GPA', data=df)
plt.title('GPA vs. Attendance Rate with Trendline')
plt.xlabel('Attendance Rate (%)')
plt.ylabel('GPA')
plt.show()

# Task 3: Bar Chart Visualization
plt.figure(figsize=(10, 6))
df['Major'].value_counts().plot(kind='bar')
plt.title('Number of Students per Major')
plt.xlabel('Major')
plt.ylabel('Number of Students')
plt.show()

# Task 4: Grouped Bar Chart
plt.figure(figsize=(10, 6))
sns.countplot(x='Graduation Year', hue='Extracurricular Activities', data=df)
plt.title('Number of Students with/without Extracurricular Activities across Graduation Years')
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.show()

# Analysis Questions
# 1. What is the most common GPA range among students?
gpa_range = pd.cut(df['GPA'], bins=[0, 1, 2, 3, 4])
most_common_gpa_range = gpa_range.value_counts().idxmax()
print("Most common GPA range:", most_common_gpa_range)

# 2. Is there any observable relationship between attendance and GPA in the scatterplot?
print("The scatterplot and trendline provide visual evidence of the relationship between attendance and GPA.")

# 3. Which major has the highest number of students enrolled?
most_common_major = df['Major'].value_counts().idxmax()
print("Major with the highest number of students enrolled:", most_common_major)

# 4. Are students with extracurricular activities more likely to have higher attendance rates?
extracurricular_attendance = df.groupby('Extracurricular Activities')['Attendance Rate (%)'].mean()
print("Average Attendance Rate by Extracurricular Activities:\n", extracurricular_attendance)
```
